```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = "english")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df["Message"]
y = df["Classification"]
```

```python
x
```

```python
y
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, train_size = 0.8)
```

```python
train_x1 =  cv.fit_transform(train_x)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(train_x1, train_y)
```

```python
test_x1 = cv.transform(test_x)
```

```python
model.score(test_x1, test_y)
```

```python
message = input("Enter a message : ")
info = cv.transform([message]).toarray()
result = model.predict(info)
if result[0] == 1:
    print("Spam")
else : 
    print("Not A Spam")
```
